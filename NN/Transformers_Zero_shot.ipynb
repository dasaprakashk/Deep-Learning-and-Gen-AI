```python
from transformers import pipeline
import sentencepiece
import pandas as pd
```



*   Text classification



```python
text = """Dear Amazon, last week I ordered an Optimus Prime action figure from your online store in Germany.
Unfortunately, when I opened the package, I discovered to my horror that I have been sent an action figure of Megtron
instead! As a lifelong enemy of Decepticons, I hope you can understand my dilemma. To resolve the issue, I demand the exchange of
Megatron for the Optimus Prime figure I ordered. Enclosed are the copies of my records concerning this purchase. I expect to hear from you soon.
Sincerely, Bumblebee."""
```

```python
classifier = pipeline("text-classification")
outputs = classifier(text)
pd.DataFrame(outputs)
```

* Named Entity Recognition

```python
ner_tagger = pipeline('ner', aggregation_strategy="simple")
outputs = ner_tagger(text)
pd.DataFrame(outputs)
```

* Question and Answering

```python
reader = pipeline("question-answering")
question = "What does the customer want?"
outputs = reader(question=question, context=text)
pd.DataFrame([outputs])
```

* Summarization

```python
summarizer = pipeline("summarization")
outputs = summarizer(text, max_length=60, clean_up_tokenization_spaces=True)
print(outputs[0]['summary_text'])
```

* Translation

```python
translator = pipeline("translation_en_to_de", model="Helsinki-NLP/opus-mt-en-de")
outputs = translator(text, clean_up_tokenization_spaces=True, min_length=100)
print(outputs[0]['translation_text'])
```

* Text Generation

```python
generator = pipeline("text-generation")
response = "Dear Bumblebee, I am sorry to hear that your order was mixed up."
prompt = text + "\n\nCustomer service response:\n" + response
outputs = generator(prompt, max_length=300)
print(outputs[0]['generated_text'])
```
